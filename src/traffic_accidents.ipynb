### (7) calculate summary statistics for the number of persons injured in all NYC and for a couple of selected boroughs (you can choose the two boroughs)

Here we will be using the counts of the column 'NUMBER OF PERSONS INJURED' to find the amount of people injured.

We will then use numpy to calculate the mean, median, percentiles (25 and 75), max, and min.

```python
# for entire NYC area
import numpy as np
nyc_injured = np.array(df['NUMBER OF PERSONS INJURED'])
print('Mean number of persons injured for NYC area is ',np.mean(nyc_injured))
print('Median number of persons injured for NYC area is ',np.median(nyc_injured))
print('The 25 percential number of persons injured for NYC area is ',np.percentile(nyc_injured, 25))
print('The 75 percential number of persons injured for NYC area is ',np.percentile(nyc_injured, 75))

print('Max number of persons injured for NYC area is ',np.max(nyc_injured))
print('Min number of persons injured for NYC area is ',np.min(nyc_injured))
```

```python
# for Manhattan area
import numpy as np
# grabbing rows where the borough column is manhattan
nyc_manhattan = df[df['BOROUGH'] == 'MANHATTAN']
# taking out the column 'number of persons injured' from above and transforming into numpy array
nyc_injured_manhattan = np.array(nyc_manhattan['NUMBER OF PERSONS INJURED'])
print('Mean number of persons injured for MANHATTAN is ',np.mean(nyc_injured_manhattan))
print('Median number of persons injured for MANHATTAN is ',np.median(nyc_injured_manhattan))
print('The 25 percential number of persons injured for MANHATTAN is ',np.percentile(nyc_injured_manhattan, 25))
print('The 75 percential number of persons injured for MANHATTAN is ',np.percentile(nyc_injured_manhattan, 75))
print('Max number of persons injured for MANHATTAN is ',np.max(nyc_injured_manhattan))
print('Min number of persons injured for MANHATTAN is ',np.min(nyc_injured_manhattan))
```

```python
# for Bronx area
# grabbing rows where the borough column is bronx
nyc_manhattan = df[df['BOROUGH'] == 'BRONX']
# taking out the column 'number of persons injured' from above and transforming into numpy array
nyc_injured_manhattan = np.array(nyc_manhattan['NUMBER OF PERSONS INJURED'])
print('Mean number of persons injured for BRONX is ',np.mean(nyc_injured_manhattan))
print('Median number of persons injured for BRONX is ',np.median(nyc_injured_manhattan))
print('The 25 percential number of persons injured for BRONX is ',np.percentile(nyc_injured_manhattan, 25))
print('The 75 percential number of persons injured for BRONX is ',np.percentile(nyc_injured_manhattan, 75))
print('Max number of persons injured for BRONX is ',np.max(nyc_injured_manhattan))
print('Min number of persons injured for BRONX is ',np.min(nyc_injured_manhattan))
```

### (9) shows the covariance between each pair of the columns

The columns that we will be choosing will be the following:

- LOCATION
- NUMBER OF PERSONS INJURED
- NUMBER OF PERSONS KILLED

Specifically, we are choosing columns that hold the location, as well as the amount of people injured/killed. 

It is important to note that we are looking for numerical columns, as it wouldn't make sense/ work entirely if we chose categorical columns.

This will hopefull give us an idea on what part of NYC has the "worst" traffic accidents.


```python
# First, we extract the longitude and latitude (together) from the dataframe
# we will then calculate the covariance of injured and killed seperately with the "location" 
# dataframe we get from the previous step, for which will have 2 columns
```

```python
# extract columns from dataframe
df_latitude = df['LATITUDE']
df_longitude = df['LONGITUDE']
# get num of persons injured
df_injured = df['NUMBER OF PERSONS INJURED']
df_killed = df['NUMBER OF PERSONS KILLED']
```

```python
np.array(df_latitude)
np.array(df_longitude)
np.array(df_injured)
np.array(df_killed)
```

```python
# latitude vs injured
lat_injured = np.cov(df_latitude,df_injured)
# latitude vs killed
lat_killed = np.cov(df_latitude,df_killed)
# longitude vs injured
long_injured = np.cov(df_longitude,df_injured)
# longitude vs killed
long_killed = np.cov(df_longitude,df_killed)
print(lat_injured)
print(lat_killed)
print(long_injured)
print(long_killed)
```

It appears to be that all of our calculations show a very small/weak covariance between our chosen columns.

## (10) which month did the most number of accidents occur

First, we will need to import the calendar module to easily extract the months of where the accidents happen, then we will make a new column to fit datetime and then make a new column of the dates.

Lastly, we will throw the month column into a counter object and plot the results.

```python
#import calendar module and extract months
import calendar
df['date'] = pd.to_datetime(df['CRASH DATE'])
df['month'] = df['date'].dt.month_name()
```

```python
month_counts = Counter(df['month'])
```

```python
month, counts = zip(*month_counts.items())
```

```python
plt.bar(month,counts)
plt.xlabel('month')
plt.ylabel('counts')
plt.title('number of accidents in each month in NYC area')
```

Looking at the bar graph above, we can see that accidents reach a lull near April. 

April is also when spring is at it's peak, we can hypothesize that it's due to more people taking the train, walking, instead of driving to avoid the cold, thus lowering the amount of vehicles on the road and lowering traffic accidents as a whole. This is further supported as for how January has the highest amount of accidents, as it is peak winter.
